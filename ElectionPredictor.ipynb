```python
sentiment=[]
for x in range(len(df)):
    negative=df['angry'][x]+df['sad'][x]
    positive=df['love'][x]+df['wow'][x]
    if (positive>negative):
        sentiment.append(1)
    elif (positive<negative):
        sentiment.append(-1)
    else:
        sentiment.append(0)
```

```python
dfPart=pd.DataFrame(data=sentiment)
df['sentiment']=dfPart
```

```python
# FInding the discrete probability distribution over positive, negative and neutral datasets
listAll=[]
listPos=[]
listNeg=[]
listNeu=[]
words=["haha","like","love","wow","angry","sad"]
for xx in words:
    totalSum=0
    partialSumNeg=0
    partialSumPos=0
    partialSumNeu=0
    for x in range(len(df)):
        totalSum=df[xx][x]+totalSum
        if df['sentiment'][x]==-1:
            #total sum of reactions for neg sentiment
            partialSumNeg=df[xx][x]+partialSumNeg
        if df['sentiment'][x]==1:
            #total sum of reactions for pos sentiment
            partialSumPos=df[xx][x]+partialSumPos
        if df['sentiment'][x]==0:
            partialSumNeu=df[xx][x]+partialSumNeu
            
    listAll.append(totalSum)
    listPos.append(partialSumPos)
    listNeg.append(partialSumNeg)
    listNeu.append(partialSumNeu)
```

```python
probabilityScorePos=[]
probabilityScoreNeg=[]
probabilityScoreNeut=[]
x=0
for x in range(len(listAll)):
        scorePos=listPos[x]/listAll[x]
        scoreNeg=listNeg[x]/listAll[x]
        scoreNeu=listNeu[x]/listAll[x]
        probabilityScorePos.append(scorePos)
        probabilityScoreNeg.append(scoreNeg)
        probabilityScoreNeut.append(scoreNeu)
```

```python
sentimentScore=[]
x=0
for x in range(len(listAll)):
    score=probabilityScorePos[x]*1+probabilityScoreNeg[x]*(-1)+probabilityScoreNeut[x]*(0)
    sentimentScore.append(score)
    
```

```python
sentimentScore
```

```python
sentDf=pd.DataFrame(data=words,columns=["sentiment"])
sentDf['sentimentScore']=sentimentScore
sentDf
```
